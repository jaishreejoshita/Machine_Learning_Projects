## To predict the salary based on the years of experience. Hence, the objective is to construct a simple linear regression model. This model will recommend an appropriate salary for a new hire based on their experience level within the company.


```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
```

```python
#loading dataset
dataset = pd.read_csv('Salary_Data.csv')
dataset
```

```python
# Splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split # this function randomly split dataset into training and testing sets according to a specified ratio, typically, with a larger portion allocating for training.
X_train, X_test, y_train, y_test = train_test_split(dataset.iloc[:, :-1].values,#selecting years of experience column
                                                    dataset.iloc[:, -1].values, #selecting salary column (second column)
                                                    test_size=0.2,
                                                    random_state=0) #radom state can be any arbitrary no. which help in getting the same split of data every time we run the code

#Note: The output of train_test_split is four variables:
#X_train contains training features,
#X_test contains the testing features,
#y_train contains the training target variable values, and
#y-test contains the testing target variable values.

# Training the simple linear regression model on the training set
regressor = lm.LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the test set results
y_pred = regressor.predict(X_test)

# Visualizing the training set results
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

#Note: Visualizing the training set results allows us to assess how effectively our model fits the training data, providing insights into its leanring process and how well it captures the patterns in the data.

# Visualizing the test set results
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue') #we usee the same trained model as it is same for the testing data set
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

#Note: Same prediction line is used for both the training and testing data plots, it porvides insight into how models predictions  learn from the training data applied to new unseen data
#Note: Visualizing the testing set results allows us to evaluate how well our model generalizes to new, unseen data.
```
